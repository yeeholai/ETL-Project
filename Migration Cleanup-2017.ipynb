```python
import pandas as pd
import numpy as np
```

```python
mig_2017 = pd.read_excel("Resources/State_to_State_Migrations_Table_2017.xls")
```

```python
# drop rows that are completely empty
mig_2017.dropna(how='all',inplace=True)
```

```python
# keep the middle portion of the table. drop header and footer comments
mig_2017 = mig_2017.iloc[3:62,:].reset_index(drop=True)
```

```python
# drop column headers that appear in the middle of the dataset
mig_2017.drop(mig_2017.index[31:34], inplace=True)
```

```python
# fill in NAs in row[1] with values from row above, ie "population 1 year and over" etc.
mig_2017.iloc[:2,:9].ffill(inplace=True)
```

```python
# drop columns that are repeating row headers, except the first occurrence, which will be used as row index
mig_2017 = mig_2017.drop(columns=mig_2017.columns[(mig_2017 == "Current residence in --").any()][1:])
```

```python
# reset index with states name
mig_2017.set_index(mig_2017.iloc[:,0].name, inplace=True)
```

```python
# drop every other column (keep Estimate columns only)
mig_2017 = mig_2017[[k for i, k in enumerate(mig_2017.columns,1) if i %2!=0]]
```

```python
# reset index in preparation for next step- renaming column headers with row[1], removing the ugly "Table with row..."
mig_2017.reset_index(inplace=True)
```

```python
# rename column headers with state names from row 1
mig_2017.rename(columns=mig_2017.iloc[1], inplace=True)
```

```python
# Remove the first 3 rows in df, which are unnecessary column headers
mig_2017= mig_2017.iloc[3:,:]

# reset index with states name
mig_2017.set_index(mig_2017.iloc[:,0].name, inplace=True)

# rename "Total" column header
mig_2017.rename(columns={"Total":"Different state of residence 1 year ago"}, inplace=True)

mig_2017
```
