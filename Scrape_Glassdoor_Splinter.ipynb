```python
from splinter import Browser
from bs4 import BeautifulSoup as bs
import requests
import time
from webdriver_manager.chrome import ChromeDriverManager
import re
import pandas as pd
```

```python
# Set up Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# empty lists for collection of job titles and wages

job_title_list = []
wage_list = []
```

```python
# Visit glassdoor salaries page
url = "https://www.glassdoor.com/Salaries"
browser.visit(url)
```

```python
# search by location using California
# Glassdoor doesn't populate search using state, so have to fill in the search with a city name. 
# Used 'Paso Robles, CA (US)' & 'San Diego, CA (US)'

browser.find_by_id("LocationSearch").fill("Paso Robles, CA (US)")
```

```python
browser.click_link_by_id("HeroSearchButton")
```

```python
# note: glassdoor prompts you to sign in with an account after hitting search button.  I manually entered my login information.
```

```python
# Glassdoor would stop populating salary after a certain number of clicks.  Manually clicked through pages and ran the code
# below to get salary from each page

html = browser.html
soup = bs(html, 'html.parser')
```

```python
job_titles = soup.find_all('span', attrs={"data-test":re.compile(r"job-title$")})
    
for i in job_titles:
    job_title_list.append(i.text)
```

```python
wage = soup.find_all('h3', attrs={"class":"m-0 css-g261rn", "data-test":None})
    
for i in wage:
    if i != "Is this salary info helpful?":
        wage_list.append(i.text)
    else:
        pass
```

```python
len(job_title_list)==len(wage_list)
```

```python
glassdoor_df = pd.DataFrame({"title":job_title_list, "wage": wage_list})
glassdoor_df
```

```python
glassdoor_df.to_csv("glassdoor_df.csv")
```
